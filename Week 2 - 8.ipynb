```python
import boto3
from trp import Document

bucketname = "kadakareer1"
tablename = "Annex-G_Psycho-Evaluation-Form 2.jpg"

module = boto3.client('textract')

response = module.analyze_document(
    Document = {
        'S3Object': {
            'Bucket': bucketname,
            'Name': tablename
        }
    }, FeatureTypes = ["TABLES"])
doc = Document(response)

n = open("Week 2 - 8.txt", "w")
x = '-----Amazon Textract has detected as follows-----'
print(x)
n.write(x)
for page in doc.pages:
    for table in page.tables:
        for r, row in enumerate(table.rows):
            itemName  = ""
            for c, cell in enumerate(row.cells):
                y = "Table[{}][{}] = {}".format(r, c, cell.text)
                print(y)
                n.write("\n" + y)
n.close()
```
