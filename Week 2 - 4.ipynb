```python
import boto3
from trp import Document

bucketname = "kadakareer1"
formname = "Appendix B - Annex C MER- AO2013-0006 1.jpg"

module = boto3.client('textract')

response = module.analyze_document(
    Document = {
        'S3Object': {
            'Bucket': bucketname,
            'Name': formname
        }
    }, FeatureTypes = ["FORMS"])
doc = Document(response)

n = open("Week 2 - 4.txt", "w")
x = '-----Amazon Textract has detected as follows-----'
print(x)
n.write(x)
for page in doc.pages:
    for field in page.form.fields:
        y = "Key: {}, Value: {}".format(field.key, field.value)
        print(y)
        n.write("\n" + y)
n.close()
```
